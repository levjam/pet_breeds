```python
def download_single_image(dest, breed):
    urls = search_images(breed, max_images=5)
    for url in urls:
        try:
            download_url(url, dest=f'{breed}.jpg', show_progress=False)
            break
        except exception as e:
            print(f'Error occured while downloading image for {breed} from {url}')
            print(f'Error: {e}')

def search_images(term, max_images=200):
    result = ddg_images(term, max_results=max_images)
    result = L(result)
    return result.itemgot('image')

dest = Path(os.getcwd(), 'pets')

for breed in dogs:
    download_single_image(dest, breed)
```

```python
dls = ImageDataLoaders.from_name_re(path,
                                    get_image_files(path/'images'),
                                    pat='(.+)_\d+.jpg',
                                    item_tfms=Resize(460),
                                    batch_tfms=aug_transforms(size=224,min_scale=0.75))
learn = vision_learner(dls, models.resnet50, metrics=accuracy)
```

```python
dls.show_batch()
```

```python
learn.fine_tune(1)
learn.path=Path('.')
learn.export()
```

```python
learn = load_learner('export.pkl')
```

```python
labels = learn.dls.vocab

def predict(img):
    img = PILImage.create(img)
    pred, pred_idx, probs = learn.predict(img)
    return {labels[i]: float(probs[i]) for i in range(len(labels))}

import gradio as gr
# gr.interface(fn=predict,
#              inputs=gr.inputs.Image(=(512, 512)),
#              outputs=gr.outputs.Label(num_top_classes=5)
#             ).launch(share=True)
gr.Interface(fn=predict, inputs=gr.Image(height=512, width=512), outputs=gr.Label(num_top_classes=3)).launch(share=True)
```
